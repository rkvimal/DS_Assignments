### Amazon Dataset

```python
## Import the relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amazon = pd.read_csv('amazon.csv')

"""amazon['discounted_price'] = amazon['discounted_price'].astype(str).str.replace(',', '').astype(float)
amazon['actual_price'] = amazon['actual_price'].astype(str).str.replace(',', '').astype(float)
amazon['discount_percentage'] = amazon['discount_percentage'].astype(str).str.replace(',', '').astype(float)
amazon['rating_count'] = amazon['rating_count'].astype(str).str.replace(',', '').astype(float)
"""

amazon['discounted_price'] = amazon['discounted_price'].str.replace(',', '').str.replace('.', '').astype(float)
amazon['actual_price'] = amazon['actual_price'].str.replace(',', '').str.replace('.', '').astype(float)
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%', '').astype(float)
amazon['rating_count'] = amazon['rating_count'].str.replace(',', '').str.replace('.', '').astype(float)

# Remove any non-numeric characters from the 'rating' column
amazon['rating'] = amazon['rating'].str.replace('[^\d.]', '', regex=True)

# Convert the 'rating' column to numeric
amazon['rating'] = pd.to_numeric(amazon['rating'])

print('Dimensions of the dataset :',amazon.shape)
print(amazon.info())
amazon.head()
```

Since the type of numerical columns is object, we will convert them to float 

```python
## check null values in the dataset
amazon.isna().sum()
```

The above output shows that there are 2 null values present in the rating_count column in the dataset

Since the product category contains delimitted values , we will split it on the basis on the delimitter and store it in a new column

```python
amazon['ParentCategory'] = amazon['category'].str.split('|',expand= True)[0].str.replace('&',' ')
amazon.head(2)
```

```python
# Calculate the mean rating for each parent category
Mean_Rating = amazon.groupby('ParentCategory')['rating'].mean().reset_index()
Mean_Rating['rating'] = Mean_Rating['rating'].round(1)
Mean_Rating
```

Q1)  What is the average rating for each product category ?

```python
Mean_Rating = amazon.groupby('ParentCategory')['rating'].mean().reset_index()

print(round(Mean_Rating,1))

plt.figure(figsize=(12, 4))

# Plot of Catergorywise Average Ratings
sns.barplot(x='ParentCategory', y='rating', data=Mean_Rating, color='skyblue')
plt.title('Average Rating of Parent Categories')
plt.xlabel('Parent Categories')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Observations:
1) From the above barplot, we can conclude that the lowest average Rating is for the Car Motorbike Category.
2) Both, OfficeProducts & Toys Games have the highest average rating

Q3) What is the distribution of discounted prices vs. actual prices ?

```python
print(amazon[['discounted_price','actual_price']].corr())
sns.scatterplot(x='discounted_price',y='actual_price',data=amazon)
plt.title('Discounted Price vs Actual Price')
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.show()
```

Observations:
From the above scatterplot, we can observe that there is positive correlation between Discounted Price and Actual Price.
However, there are some extreme values which are impacting the correlation coefficient. 

#### Self Calculations to explore relationship excluding the extreme values

```python
# Calculate the 95th percentile of both columns
discounted_price_95th = amazon['discounted_price'].quantile(0.95)
actual_price_95th = amazon['actual_price'].quantile(0.95)

# Winsorize the columns
amazon['discounted_price_winsorized'] = amazon['discounted_price'].clip(upper=discounted_price_95th)
amazon['actual_price_winsorized'] = amazon['actual_price'].clip(upper=actual_price_95th)

# Calculate the correlation coefficient
correlation = amazon[['discounted_price_winsorized', 'actual_price_winsorized']].corr()
print(correlation)

sns.scatterplot(x='discounted_price_winsorized',y='actual_price_winsorized',data=amazon)
plt.title('Winsorized Discounted Price vs Winsorized Actual Price')
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.show()
```

Q6) What are the most popular product keywords ?

```python
from collections import Counter
import re
from nltk.corpus import stopwords

# Download NLTK stopwords
import nltk
nltk.download('stopwords')

# Get English stopwords
stop_words = set(stopwords.words('english'))

# Tokenize and count the frequency of each word
keywords_counter = Counter()
for name in amazon['product_name']:
    words = re.findall(r'\b\w+\b', name.lower())
    words = [word for word in words if word not in stop_words and len(word) > 2] ## removed one letter words which remained even after removing stop words
    keywords_counter.update(words)

# Sort keywords by frequency in descending order
popular_keywords = keywords_counter.most_common()

# Display the most popular keywords
popular_keywords[:10]  # Display top 10 keywords
```

Q7)  What are the most popular product reviews ?

```python
# Combine review_title and review_content into a single text column
amazon['review_text'] = amazon['review_title'] + ' ' + amazon['review_content']

# Tokenize and count the frequency of each word
review_counter = Counter()
for text in amazon['review_text']:
    words = re.findall(r'\b\w+\b', str(text).lower())
    words = [word for word in words if word not in stop_words and len(word) > 2]
    review_counter.update(words)

# Sort reviews by frequency in descending order
popular_reviews = review_counter.most_common()

# Display the most popular reviews
popular_reviews[:10]  # Display top 10 reviews
```

Q4)  Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the  top 5 longest songs among the tracks of the top 5 popular artists using BarPlot.

```python

# Calculate the average popularity for each artist
artist_popularity = spotify.groupby('Artist')['Popularity'].mean().reset_index()

# Find the top 5 popular artists
top_popular_artists = artist_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Filter the dataset to include only tracks by the top 5 popular artists
top_tracks = spotify[spotify['Artist'].isin(top_popular_artists['Artist'])]

# Find the top 5 longest songs among the tracks of the top 5 popular artists
top_longest_songs = top_tracks.sort_values(by='Duration (ms)', ascending=False).head(5)

# Represent the duration of the top 5 longest songs using a bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_longest_songs['Track Name'], top_longest_songs['Duration (ms)'], color='skyblue')
plt.xlabel('Track Name')
plt.ylabel('Duration (ms)')
plt.title('Top 5 Longest Songs Among Tracks of Top 5 Popular Artists')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
